<p style="font-family: 'Times New Roman'; font-size: 18px;">
Your program should use pandas to create a Series where the days are used as indices, and the corresponding temperature are the data points. After creating the Series, print the resulting data structure.
</p>
<span style="font-family: 'Times New Roman'; font-size: 18px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(i) Retrieve and print the temperature for Tuesday.
</span><br>
<span style="font-family: 'Times New Roman'; font-size: 18px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(ii) Slice and display the temperatures from Wednesday to Saturday.
</span><br>
<span style="font-family: 'Times New Roman'; font-size: 18px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(iii) Calculate and print the average temperature for the week.
</span>

```python
import pandas as pd

# Create a pandas Series with days as indices and temperature as data
temperature_data = {'Mon': 25, 'Tue': 28, 'Wed': 22, 'Thu': 30, 'Fri': 24, 'Sat': 27, 'Sun': 31}
temperature_series = pd.Series(temperature_data)

# Print the resulting Series
print("Temperature Data:")
print(temperature_series)

# (i) Retrieve and print the temperature for Tuesday
print("\nTemperature for Tuesday:", temperature_series['Tue'])

# (ii) Slice and display the temperatures from Wednesday to Saturday
print("\nTemperatures from Wednesday to Saturday:")
print(temperature_series['Wed':'Sat'])

# (iii) Calculate and print the average temperature for the week
average_temperature = temperature_series.mean()
print("\nAverage Temperature for the Week:", round(average_temperature, 2))
```

2. You are given dictionary lists of Name, countries, and age as follows:

    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<i><strong>Age</strong></i>: <i>28</i>, <i>34</i>, <i>29</i>, <i>33</i>, <i>25</i>, <i>40</i>, <i>30</i>, <i>26</i> <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<i><strong>Country</strong></i>: <i>Italy</i>, <i>Germany</i>, <i>American</i>, <i>Ireland</i>, <i>Finland</i>, <i>France</i>, <i>China</i>, <i>England</i>
</span><br><br>

<span style="font-family: 'Times New Roman'; font-size: 18px;">
Create an indexed DataFrame where indices of the DataFrame are <i>‘rank1’</i>, <i>‘rank2’</i>, …, <i>‘rank8’</i>.
</span><br>

<span style="font-family: 'Times New Roman'; font-size: 18px;">
(a) Add a new row with a suitable name, age, and country to the DataFrame.<br>
(b) Index the row labelled <i>‘rank2’</i> and rename it to <i>‘rank3’</i>.<br>
(c) Index the rows labelled <i>‘rank8’</i> and <i>‘rank9’</i>.<br>
(d) Index the column labelled <i>‘Name’</i> and <i>‘Country’</i>.
</span>

```python
import pandas as pd

# Given data
data = {'Name': ['Tom', 'Jack', 'Steve', 'Ricky', 'Mike', 'John', 'Chan', 'William'],
        'Age': [28, 34, 29, 33, 25, 40, 30, 26],
        'Country': ['Italy', 'Germany', 'American', 'Ireland', 'Finland', 'France', 'China', 'England']}

# Create a DataFrame with custom indices
df = pd.DataFrame(data, index=['rank1', 'rank2', 'rank3', 'rank4', 'rank5', 'rank6', 'rank7', 'rank8'])

# (a) Add a new row
new_row_data = {'Name': 'NewPerson', 'Age': 31, 'Country': 'Spain'}
df.loc['rank9'] = new_row_data

# (b) Rename the row labelled 'rank2' to 'rank3'
df = df.rename(index={'rank2': 'rank3'})

# (c) Index the rows labelled 'rank8' and 'rank9'
selected_rows = df.loc[['rank8', 'rank9']]

# (d) Index the column labelled 'Name' and 'Country'
selected_columns = df[['Name', 'Country']]

# Display the modified DataFrame and selected rows/columns
print("Modified DataFrame:")
print(df)

print("\nSelected Rows (rank8 and rank9):")
print(selected_rows)

print("\nSelected Columns (Name and Country):")
print(selected_columns)
```

3. Write a Python program using Matplotlib to create both a pie chart and a bar chart representing the popularity of programming languages. Ensure that both charts include labels, percentages, and a colour scheme. Add a title to each chart, such as “Programming Language Popularity”. Explode the most popular programming language in both charts based on the provided data.
</span><br>

<span style="font-family: 'Times New Roman'; font-size: 18px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;languages = [‘C’, ‘C++’, ‘Java’, ‘Python’, ‘PHP’]<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;students = [23, 17, 35, 29, 12]
</span>

```python
import matplotlib.pyplot as plt

# Data
languages = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]

# Explode the most popular language (e.g., Python)
explode = [0, 0, 0.1, 0, 0]

# Pie Chart
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(students, labels=languages, autopct='%1.1f%%', startangle=90, explode=explode, colors=plt.cm.Paired.colors)
plt.title('Programming Language Popularity (Pie Chart)')

# Bar Chart
plt.subplot(1, 2, 2)
plt.bar(languages, students, color=plt.cm.Paired.colors)
plt.xlabel('Programming Languages')
plt.ylabel('Number of Students')
plt.title('Programming Language Popularity (Bar Chart)')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
